# Plot-Fugure, Axes에 대한 이야기
- 플롯(plot)은 figure와 axes로 구성
- plt.Fugure: 축, 그래프, 텍스트, 레이블 등 ... 모든 객체를 포함하는 컨테이너, 그림 그자체
- Plt.Axes: 축, 눈금, 레이블, 제목 등의 요소들이 모인 그래프 (axis와는 다르다)
  - axes는 하나의 그래프, axis는 축
  - 사전적으로는 axes는 axis의 복수형
  - matplotlib에서의 axes는 그래프, axis는 축
  - 관습적으로 axes객체를 ax라는 변수에 저장
  - axes는 subplot 으로 사용된다.
- Figure 안에 한개나 여러개의 axes가 들어 있을 수 있다.
- Plt로 그래프를 그리면 figure, axes는 자동으로 생성된다.

# Multi Subplot

```python
ax1 = plt.axes()
ax2 = plt.axes([0.5, 0.5, 0.1, 0.2])
```

- plt.subplot(nrows, ncols, index) 하나의 subplot 생성
- 간단한 subplot 배치에 유용
- subplot은 axes

```python
for i in range (1, 10):
  plt.subplot(3, 3, i) # 3행 3열 i번째 위치에 플롯 붙이기
  plt.text(0.5, 0.5, str((3, 3, i)), fontsize = 18, ha = 'center') # ha는 수평정렬, va는 수직정렬
```

```python
# 위에 동일하지만 figure을 직접 생성하여 만들면,
fig = plt.figure()
for i in range(1, 10):
  plt.subplot(3, 3, i)
  plt.text(0.5, 0.5, str((3, 3, i)), fontsize = 10, ha = 'center')
fig.subplots_adjust(hspace=0.5, wspace=0.5)
```

- fig, ax = plt.subplots(3, 3, sharex = 'col', sharey = 'row', .....)
- 여러개의 subplot을 동시에 생성

```python
fig, ax = plt.subplots(3, 3, sharex = 'col', sharey = 'row')
(fig, ax)
```

```python
for i in range(3):
  for j in range(3):
    ax[i, j].text(0.5, 0.5, str((i, j)), ha='center') # subplot들은 2차원 ndarray 형태로 ######################################
fig
```

```python
grid = plt.GridSpec(2, 3, wspace=1, hspace=1)
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2])
```

```python
# subplot의 일반적인 사용 예

plt.figure(figsize = (5, 6)) # 5인치 6인치 // 100dpi 기준 인치...
x = range(1, 21)
columns = [np.random.randn(20)* i for i in range(1, 7)]

i = 0
for c in columns:
  i += 1
  plt.subplot(3, 2, i)
  plt.plot(x, c)
```

#contour, contourf

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)
plt.contourf(X, Y, Z, levels = 10, cmap = 'Blues') #coutour는 선, contourf는 면
plt.colorbar()
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
C = plt.contour(X, Y, Z, levels = 10, cmap = 'rainbow') #coutour는 선, contourf는 면
plt.clabel(C, inline=True) #등고선 위에 수치 표시
plt.colorbar()
```

#Pie

```python
data = [1, 2, 3, 4, 5]
plabel = ['a', 'b', 'c', 'd', 'e']
plt.pie(data, labels = plabel, autopct = '%.2f%%')
plt.legend(plabel)
```

```python
data = [1, 2, 3, 4, 5]
plabel = ['a', 'b', 'c', 'd', 'e']
margin = [0.1, 0.1, 0.5, 0.1, 0.1]  # 중심에 대한 거리
plt.pie(data, explode = margin, labels = plabel, autopct = '%0.2f%%')
plt.legend(plabel)
```

#3D

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
x = range(1, 101)
y = np.random.normal(500, 1, 100)
z = np.random.randn(100)
ax = fig.add_subplot(111, projection='3d')  # axes를 3차원으로 설정
ax.scatter(x, y, z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(np.sqrt(X**2 + Y**2))
ax.contour3D(X, Y, Z, 200, cmap='Blues')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
