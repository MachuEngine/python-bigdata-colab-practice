Concat

```python
df1 = pd.DataFrame({'a':[14, 55], 'b':[64, 21]}, index = [0, 1])
df2 = pd.DataFrame({'a':[45, 83], 'b':[12, 76]}, index = [0, 1])
df3 = pd.DataFrame({'c':[64, 43], 'd':[67, 96]}, index = [0, 1])
df4 = pd.DataFrame({'c':[61, 12], 'd':[94, 34]}, index = [2, 3])
df5 = pd.DataFrame({'c':[14, 74], 'd':[27, 61]}, index = [10, 34])
df6 = pd.DataFrame({'a':[45, 33], 'b':[24, 53]}, index = [2, 3])

print (df1);print ()
print (df2);print ()
print (df3);print ()
print (df4);print ()
print (df5);print ()
print (df6);print ()
```

```python
# 기본 axis = 0, row행으로 붙인다.
print (pd.concat([df1, df2])) # column이 같고, index가 같은 경우
print (pd.concat([df1, df3])) # column이 다르고, index가 같은 경우
print ('-----------')
print (pd.concat([df1, df4], verify_integrity = True)) # 오버랩 없음
print ()
print (pd.concat([df1, df3])) # 오버랩 되도 그대로 붙인다.
```

```python
#print (pd.concat([df1, df3], verify_integrity = True)) # index 오버랩 있으면 에러
print (pd.concat([df1, df6], verify_integrity = True)) # column 오버랩은 신경안쓴다.
print ()
print (pd.concat([df1, df3], ignore_index = True)) # 오버랩 있든 말든 무시하고 새로운 index 생성
print (pd.concat([df1, df5], ignore_index = True)) # 오버랩 있든 말든
```
