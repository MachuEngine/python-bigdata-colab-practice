### Google Drive 마운트

```python
# [데이터 수집]

f = open('./gdrive/MyDrive/Colab Notebooks/ch4_temperature/temperature.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')

# 헤더 건너 뛰기
header = next(data)  # ['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']

high_temp = []  # 최고기온 리스트 초기화
low_temp = []   # 최저기온 리스트 초기화
years = []      # 연도 리스트 초기화

# 데이터 수집 시작
for row in data:
  # 빈 행은 건너뛰기
  if not row:
    continue

  # 컬럼 개수가 기대치(여기선 최소 5개)보다 작으면 건너뛰기
  if len(row) < 5:
    continue

  # 날짜 데이터 -로 split한 후 date에 저장
  try:
    year, month, day = map(int, row[0].split('-'))
  except (ValueError, AttributeError):
    # row[0]이 "YYYY-MM-DD" 가 아닐 때
    continue

  # 조건: 1980년 이후, 1월 1일만 필터링
  if year >= 1980 and month == 1 and day == 1:
    # 최저 기온, 최고 기온 데이터 실수로 변환
    try:
      row[-1] = float(row[-1])
      row[-2] = float(row[-2])
    except ValueError:
      continue

    years.append(year) # 1980년 이후, 1월 1일인 경우, 해당 년도 데이터를 추가
    high_temp.append(float(row[-1])) # 1980년 이후, 1월 1일인 경우 해당 최고기온 데이터를 추가
    low_temp.append(float(row[-2]))  # 1980년 이후, 1월 1일인 경우 해당 최저기온 데이터를 추가

# 데이터 수집 종료
f.close()
```

### 메인 동작

1. CSV 파일 열기
- `cp949` 인코딩으로 읽어옴

2. 헤더 제거
- 첫 줄(컬럼명) 은 데이터가 아니므로 스킵

3. 루프별 검증
- 빈 행 또는 컬럼 수가 5개 미만인 행 건너뜀
- 날짜 칼럼을 파싱하지 못하면 건너뜀

4. 필터링
- 연도가 1980 이상이고, 월과 일이 1월 1일인 경우만 처리

5. 기온 값 변환 및 저장
- 문자열 형태의 최저·최고 기온을 float 로 변환
- `years`, `high_temp`, `low_temp` 리스트에 추가

6. 파일 종료
- `f.close()` 로 파일 핸들 정리

```python
# [데이터 시각화]

# 폰트 파일 경로 지정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
# 폰트를 Matplotlib 전역에 추가
fm.fontManager.addfont(font_path)
# 폰트 내부 이름 확인 (optional)
font_prop = fm.FontProperties(fname=font_path)
# rcParams 에서 이 이름을 기본 폰트로 설정
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

plt.figure(figsize=(12, 6))
# 최고 기온
plt.plot(years, high_temp, 'r-', label='최고기온')
# 최저 기온
plt.plot(years, low_temp, 'b-', label='최저기온')
# 그래프 제목 및 축 레이블 설정
plt.title('1980년 이후 1월 1일의 최고/최저기온 변화')
plt.xlabel('연도')
plt.ylabel('기온(℃)')
plt.legend()
plt.grid(True)
plt.show()
```
