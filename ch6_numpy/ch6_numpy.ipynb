boolean

```python
a1 = np.arange(1, 10).reshape(3, 3)
print(a1)
print(a1 > 2)
print(a1 < 7)
print(a1 < 7)
print(a1 <= 7)
print(a1 >= 7)
print(a1 == 7)
print(a1 != 7)
print('--------------')
print((a1 > 2) & (a1 < 7)) # and
print(a1[(a1 > 2) & (a1 < 7)]) # boolean 인덱싱 // 중요!!!!!!!!!!!!!!!!!
print((a1 > 2) | (a1 < 7)) # or
print(a1[(a1 > 2) | (a1 < 7)]) # boolean 인덱싱 // 중요!!!!!!!!!!!!!!!!!
print((a1 > 2) ^ (a1 < 7)) # xor
print(a1[(a1 > 2) ^ (a1 < 7)]) # boolean 인덱싱 // 중요!!!!!!!!!!!!!!!!!
print(~(a1 > 2)) # not
print(a1[~(a1 > 2)]) # boolean 인덱싱 // 중요!!!!!!!!!!!!!!!!!
```
